```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.figure_factory as ff
import plotly.io as pio
pio.templates.default = "plotly_white"
import plotly.colors as colors

```

```python
df = pd.read_csv('Superstore.csv', encoding = "latin-1")
```

# **Converting Date**

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df["Order Date"].head(5)
```

# **Creating New coloumn**


```python
df["order month"]=df["Order Date"].dt.month
df["order year"]=df["Order Date"].dt.year
df["order day week"]=df["Order Date"].dt.dayofweek
```

```python
df.head()
```

# Monthly Sales Analysis


```python
df.columns
```

```python
sales_by_month = df.groupby("order month")[["Sales"]].sum().reset_index()
```

```python
px.bar(sales_by_month, x="order month", y="Sales", title="Monthly Sales Analysis")
```

```python
## here we can see that november is the best month for sales
## sales amount = $352.61k
```

# **Sales by Category**

```python
sales_by_Category = df.groupby('Category')["Sales"].sum().reset_index()
```

```python
fig = px.pie(sales_by_Category, names="Category", values ="Sales", title="Sales by Category",
       hole = 0.5)
fig.update_traces(textposition='inside', textinfo='percent+label')
```

```python
## most sold category is technology($836.15k)
## and least sold category is Office Supplies($719k)
```

# **Sales by Sub Category**

```python
Sales_by_SubCategory = df.groupby("Sub-Category")["Sales"].sum().reset_index()
```

```python
fig = px.bar(Sales_by_SubCategory, x="Sub-Category", y="Sales", title="Sales by Sub-Category")
fig.update_layout(xaxis_tickangle=-45)
fig.show()

```

```python
## most sold Sub category is Phones($330k)
## and least sold Sub category is Fastners($3024)
```

# **Monthly Profit**

```python
df.columns
```

```python
Profit_ana = df.groupby('order month')[['Sales','Profit']].sum().reset_index()
```

```python
px.bar(Profit_ana, x='order month', y=["Profit"], title="Monthly Profit")
```

```python
Profit_ana = df.groupby('Category')[['Sales','Profit']].sum().reset_index()
```

```python
px.pie(Profit_ana,names = "Category",values = "Profit",
       hole = 0.5,title = "Profit by Category")\
.update_traces(textposition='inside', textinfo='percent+label')
```

```python
# most profitable Category is Technology($145k) & least is Furniture($18k)
```

```python
profit_ana = df.groupby('Sub-Category')[['Sales']].sum().reset_index()
px.line(profit_ana, x="Sub-Category", y = "Sales",title="Profit by Sub-Category")
```

```python
# most profitable Sub Category is Chairs($328k) & least is Fastners($3024)
```

# **Sales Analysis for customer Segment**

```python
profit_ana = df.groupby('Segment')[['Sales','Profit']].sum().reset_index()
px.bar(profit_ana, x='Segment', y = ['Sales','Profit'],title="Sales & Profit by Segment")
```

```python
color_palette = colors.qualitative.Pastel
fig = go.Figure()
fig.add_trace(go.Bar(
    x=profit_ana['Segment'],
    y=profit_ana['Sales'],
    name='Sales',
    marker_color=color_palette[0]
))
fig.add_trace(go.Bar(
    x=profit_ana['Segment'],
    y=profit_ana['Profit'],
    name='Profit',
    marker_color=color_palette[1]
))
fig.update_layout(barmode='group', title='Sales & Profit by Segment')
fig.show()
```

```python
## consumer is the most sold and profitable  Customer segment
```
